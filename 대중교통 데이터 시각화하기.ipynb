```python
import csv
f=open('./Data/subwaytime1.csv')
data=csv.reader(f)
header=next(data)
header=next(data)

n=[0] * 24
n_s=[''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a= row[j*2+4]
        if a > n[j]:
            n[j] = a
            n_s[j]= row[3] +'(' + str(j+4) + ')'    
                        
print(n, n_s)

import matplotlib.pyplot as plt
plt.bar(range(24), n)
plt.xticks(range(24), n_s, rotation=90)
plt.show()
```

```python
import csv
f=open('./Data/subwaytime1.csv')
data=csv.reader(f)
header=next(data)
header=next(data)

d=[0] * 24
d_s=[''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a= row[5+j*2]
        if a > d[j]:
            d[j] = a
            d_s[j]= row[3] +'(' + str(j+4) + ')'    
                        
print(d, d_s)

import matplotlib.pyplot as plt
plt.bar(range(24), d)
plt.xticks(range(24), d_s, rotation=90)
plt.show()
```

```python
import csv
f=open('./Data/subwaytime1.csv')
data=csv.reader(f)
header=next(data)
header=next(data)

s_in=[0] *24
s_out=[0] *24

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row [4+i*2]
        s_out[i] += row [5+i*2]
        
import matplotlib.pyplot as plt

plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = "승차")
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```
