```python
import csv
import matplotlib.pyplot as plt

f= open ('./data/Seoul.csv')
data = csv.reader(f)
next(data)

high = [] 
low = []

for row in data:
    if row[-1] != '' and row[-2] != '' :
        if int(row[0].split('-')[0]) >= 1983:
            if row[0].split('-')[1] == '10' and row[0].split('-')[2]== '26':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
            
plt.figure(figsize = (7,3))
plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')
plt.title('내 생일의 기온 변화 그래프')
plt.legend(loc = 5)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f= open ('./data/Seoul.csv')
data = csv.reader(f)
next(data)

avg= []
for row in data:
    if row[-1] != '' and row[-2] != '' :
        row[-1] = float(row[-1])
        row[-2] = float(row[-2])
        if row[0].split('-')[1] == '02' and row[0].split('-')[2]== '14':
            avg.append((row[-1]+row[-2])/2)
                
plt.figure(figsize = (7,3))
plt.plot(avg, 'green', label = "Average")
plt.title('내 생일의 평균 기온 변화 그래프')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open ('./data/Seoul.csv')
data=csv.reader(f)
header = next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '' and row[0] != '':
        if month == "08":
            aug.append(float(row[-1]))
        if month == "01":
            jan.append(float(row[-1]))
        
plt.hist(aug, bins=100, color='r', label ="JAN")
plt.hist(jan, bins=100, color='b', label = "AUG")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open ('./data/Seoul.csv')
data=csv.reader(f)
header = next(data)
day = []

for i in range(1,32):
    day.append([])

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == "08":
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
        
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi = 300)
plt.boxplot(day, showfliers=False)
plt.show()
```
