```python
import pandas as pd

data1 = {'s0' : ['홍길동', '홍이동', '홍삼동'],
        '수학' : [70,80,60],
        '과학' : [90,60,90],
        '영어' : [80, 75,75]}

df = pd.DataFrame(data1)
df.set_index('s0', inplace=True)
print(df)

a= df.loc['홍길동', '수학']
b= df.loc['홍길동', '과학']
c= df.loc['홍길동', '영어']
d= df.loc['홍이동', '수학']
e= df.loc['홍이동', '과학']
f= df.loc['홍이동', '영어']
g= df.loc['홍삼동', '수학']
h= df.loc['홍삼동', '과학']
i= df.loc['홍삼동', '영어']

j = a+d+g
k = b+e+h
l = c+f+i


data2 = {'s0' : ['홍길동', '홍이동', '홍삼동'],
        '수학' : [70,80,60],
        '과학' : [90,60,90],
        '영어' : [80, 75,75],
        '합계' : [j, k, l],
        '평균' : [int(j)/3, int(k)/3, int(l)/3]}



df1 = pd.DataFrame(data2)
df1.set_index('s0', inplace=True)

print(df1)
print('.\n')
print(df1.T)
```
