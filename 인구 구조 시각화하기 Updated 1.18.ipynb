```python
import csv
f = open('./Data/gender3.csv')
data=csv.reader(f)

m=[]
f=[]

for row in data:
    if '신도림'in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3]))
            f.append(int(row[-(i+1)]))
          
f.reverse()
print(m)
print(f)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label="남성")
plt.barh(range(101), f, label="여성")
plt.legend()
plt.show()
```

```python
import csv
f=open('./Data/gender3.csv')
data=csv.reader(f)

m=[]
f=[]

name=input('찾고 싶은 지역의 이름을 알려주세요')
for row in data:
    if '제주특별자치도' in row[0]:
        for i in row[3:104]:
            m.append(-int(i.replace(',','')))
        for i in row[106:]:
            f.append(int(i.replace(',','')))
        break

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+' 지역의 남녀 성별 인구 분포')
plt.barh(range(101), m, label="남성")
plt.barh(range(101), f, label="여성")
plt.legend()
plt.show()
```

# 제주도의 성별 인구 비율 표현하기

```python
import csv

f=open('./Data/gender3.csv')
data=csv.reader(f)

size=[]

name=input('찾고 싶은 지역의 이름을 알려주세요: ')
for row in data:
    if name in row[0]:
        m=0
        f=0
        for i in range(101):
            m += int(row[i+3].replace(',',''))
            f += int(row[i+106].replace(',',''))
        break
        
size.append(m)
size.append(f)
print(size)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels=['남','여'], autopct='%.1f%%', colors=color, startangle=90)
plt.title(name+ '지역의 남녀 성별 비율')
plt.show()
```

# 꺽은 선 그래프

```python
import csv
f=open('./Data/gender3.csv')
data = csv.reader(f)

m= []
f=[]

name=input('궁금한 동네를 입력해주세요: ')
for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for i in row[106:]:
            f.append(int(i))
    
print(m)
print(f)

import matplotlib.pyplot as plt
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

# 산점도로 표현하기

```python
import csv
f=open('./Data/gender3.csv')
data=csv.reader(f)

m=[]
f=[]

for row in data:
    if '제주특별자치도' in row[0]:
        for i in range(101):
            m.append(int(row[i+3].replace(',','')))
            f.append(int(row[i+106].replace(',','')))
        break
    
print(m)
print(f)

import matplotlib.pyplot as plt
plt.scatter(m,f, c=range(101), alpha=0.5)
plt.colorbar()
plt.title('제주도의 연령대별 성별 비율을 산점도로 표현하기')
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```
