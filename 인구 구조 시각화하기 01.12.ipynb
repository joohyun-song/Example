```python
import csv
import matplotlib.pyplot as plt

f=open('./Data/Sta.csv')
data=csv.reader(f)
header=next(data)

age=[]

for row in data:
    if row[0] == "경상북도 구미시 상모사곡동(4719064500)":
        for i in row[3:]:
            age.append(int(i))
            
print(age)

plt.plot(age)
plt.title('상모사곡동의 인구 구조')
plt.show()
```
